## Import Pandas

```python
import pandas as pd
```

## Mengenal DataFrame

Membuat DataFrame sederhana

```python
pd.DataFrame(data={'Minuman': ['Bandrek', 'Kopi Hitam'], 
                   'Harga': [4500, 3200]})
```

Mengenal index pada DataFrame

```python
pd.DataFrame(data={'Budi': ['suka', 'suka'], 
                   'Wati': ['tidak suka', 'suka.']},
             index=['Mangga', 'Jambu'])
```

## Mengenal Series

```python
pd.Series(data=[100, 150, 200, 250, 300])
```

```python
pd.Series(data=[3000, 3500, 40000], index=['Mangga', 'Pisang', 'Jambu'], name='Harga Buah')
```

## Menyimpan DataFrame sebagai File CSV

```python
minuman_df = pd.DataFrame(data={'Minuman': ['Bandrek', 'Kopi Hitam'], 
                                'Harga': [4500, 3200]})
minuman_df
```

```python
minuman_df.to_csv('./output/minuman.csv')
```

Menyimpan ke file CSV tanpa menyertakan index

```python
minuman_df.to_csv('./output/minuman.csv', index=False)
```
